```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('../Data/States.csv', header = [0,1], index_col = 0)
df2 = pd.read_csv('../Data/RobustStates.csv', header = [0,1], index_col = 0)
df3 = pd.read_csv('../Data/Districts.csv', header = [0,1], index_col = 0)
df4 = pd.read_csv('../Data/RobustDistricts.csv', header = [0,1], index_col = 0)

m = ['Pre' for i in range(3)]
MIR = pd.DataFrame({
    'Year' : [int(i) for i in range(2020, 2023)],
    'Monsoon' : m
})
MIR = pd.MultiIndex.from_frame(MIR)
m = ['Post' for i in range(3)]
MIO = pd.DataFrame({
    'Year' : [int(i) for i in range(2020, 2023)],
    'Monsoon' : m
})
MIO = pd.MultiIndex.from_frame(MIO)
ml1 = pd.read_csv('../Data/PreMonsoon_states.csv', index_col = 1)
del ml1['Unnamed: 0']
ml1 = ml1.dropna(how = 'all').iloc[:, -3:]
ml1.columns = MIR
ml2 = pd.read_csv('../Data/PostMonsoon_states.csv', index_col = 1)
del ml2['Unnamed: 0']
ml2 = ml2.dropna(how = 'all').iloc[:, -3:]
ml2.columns = MIO
ml3 = pd.read_csv('../Data/PreMonsoon_districts.csv', index_col = 0)
ml3 = ml3.dropna(how = 'all').iloc[:, -3:]
ml3.columns = MIR
ml4 = pd.read_csv('../Data/PostMonsoon_districts.csv', index_col = 0)
ml4 = ml4.dropna(how = 'all').iloc[:, -3:]
ml4.columns = MIO
ml5 = pd.read_csv('../Data/PreMonsoonR_districts.csv', index_col = 1)
ml5 = ml5.dropna(how = 'all').iloc[:, -3:]
ml5.columns = MIR
ml5 = ml5.dropna()
ml5.index = pd.read_csv('../Data/PostMonsoonR_districts.csv', index_col = 0).index.dropna()
ml6 = pd.read_csv('../Data/PostMonsoonR_districts.csv', index_col = 0)
ml6 = ml6.dropna(how = 'all').iloc[:, -3:]
ml6.columns = MIO
ml6.index = pd.read_csv('../Data/PostMonsoonR_districts.csv', index_col = 0).index.dropna()
```

```python
for i in range(3):
    df1.insert(
        column = ml1.iloc[:, i].name,
        value = ml1.iloc[:, i].values,
        loc = df1.shape[1]
    )
    df1.insert(
        column = ml2.iloc[:, i].name,
        value = ml2.iloc[:, i].values,
        loc = df1.shape[1]
    )
for i in range(3):
    df2.insert(
        column = ml1.iloc[:, i].name,
        value = ml1.iloc[:, i].values,
        loc = df2.shape[1]
    )
    df2.insert(
        column = ml2.iloc[:, i].name,
        value = ml2.iloc[:, i].values,
        loc = df2.shape[1]
    )
for i in range(3):
    df3.insert(
        column = ml3.iloc[:, i].name,
        value = ml3.iloc[:, i].values,
        loc = df3.shape[1]
    )
    df3.insert(
        column = ml4.iloc[:, i].name,
        value = ml4.iloc[:, i].values,
        loc = df3.shape[1]
    )
for i in range(3):
    df4.insert(
        column = ml5.iloc[:, i].name,
        value = ml5.iloc[:, i].values,
        loc = df4.shape[1]
    )
    df4.insert(
        column = ml6.iloc[:, i].name,
        value = ml6.iloc[:, i].values,
        loc = df4.shape[1]
    )
```

```python
df1.to_csv('../Data/States.csv')
df2.to_csv('../Data/RobustStates.csv')
df3.to_csv('../Data/Districts.csv')
df4.to_csv('../Data/RobustDistricts.csv')
```
