# Data Cleaning
Data downloaded from the wris site needs to be cleaned and placed into a single final dataframe.<br>
Firstly, let's try cleaning a single dataframe and then making that into a function to do it for all the other similar datasets.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(
    io = '../Data/groundwater_Report of Trends of Water Level_ArunachalPradesh.xls',
    sheet_name = 1,
    skiprows = 4,
    index_col = [1,2,3,4]
)
df = df.replace(' ', np.nan)
del df['S.NO.']
del df['Site ID']
m = []
for i in range(52):
    if i%2 == 0:
        m.append('Pre')
    else:
        m.append('Post')
MI = pd.DataFrame({
    'Year' : [int(i) for i in np.arange(1994, 2020, 0.5)],
    'Monsoon' : m
})
MI = pd.MultiIndex.from_frame(MI)
df.columns = MI
df = df.iloc[1:]
df.head()
```

The above code block solves our current cleaning problem for one specific state which needs to be written into a function so that all the other datasets can also be placed into our final dataframe.

```python
final_df = pd.DataFrame()
def clean_combine(loc):
    df = pd.read_excel(
        io = loc,
        sheet_name = 1,
        skiprows = 4,
        index_col = [1,2,3,4]
    )
    df = df.replace(' ', np.nan)
    del df['S.NO.']
    del df['Site ID']
    m = []
    for i in range(52):
        if i%2 == 0:
            m.append('Pre')
        else:
            m.append('Post')
    MI = pd.DataFrame({
        'Year' : [int(i) for i in np.arange(1994, 2020, 0.5)],
        'Monsoon' : m
    })
    MI = pd.MultiIndex.from_frame(MI)
    df.columns = MI
    df = df.iloc[1:]
    global final_df
    final_df = final_df.append(df)
```

The above code block creates an empty dataframe and a function that adds data onto the dataframe once ran using the file location.

We have a list of all the file locations in `loc` from which we'll also have to identify the states which have no data before we dive into creating the final dataset.

```python
No_Data = list()
j = 0
for i in range(len(loc)):
    if pd.read_excel(loc[i - j]).iloc[5,0] == 'NO DATA AVAILABLE':
        No_Data.append(loc.pop(i - j))
        j += 1
No_Data
```

Now that we've extracted all the empty datasets, we can append the remaining to our `final_df` through the function we made.
